```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./PDAC_data/clinical_data.csv')
df.drop(["histology_diagnosis", "additiol_pathologic_findings", "cause_of_death"], axis=1, inplace=True)
df = df.dropna(subset=['follow_up_days', 'vital_status', 'tumor_stage_pathological'])
# df = df[['follow_up_days', 'vital_status', 'is_this_patient_lost_to_follow_up']]
df.shape
```

```python
df.head()
```

```python
sts_days = 300
lts_days = 1000

sts = df[(df['follow_up_days'] < sts_days) & (df['vital_status'] == 'Deceased')].copy()
lts = df[(df['follow_up_days'] >= lts_days)].copy()

sts['survival_group'] = 'STS'
lts['survival_group'] = 'LTS'

df = pd.concat([sts, lts])
df.survival_group.value_counts()
```

```python
df.to_csv('./PDAC_data/clinical_data_sts_lts.csv', index=False)
```
